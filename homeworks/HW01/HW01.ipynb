1. Импорт библиотеки NumPy с общепринятым сокращением (`import numpy as np`).

```python
import numpy as np
```

2. Примеры создания массивов:
   - из обычных списков;
   - с использованием функций `zeros`, `ones`, `arange` или аналогичных.

```python
a = np.array([[1, 6, 8],
             [7, -4, 2]]) #создание двумерного массива (матрицу) 2 на 3
b = np.zeros((2,4,1)) # создание трехмерного массива 2 на 4 на 1 из нулей
c = np.arange(11, 17, 0.5) #создание строки с элементами от 11 до 17, не включая, и с шагом 0.5 
d = np.linspace(23, 28, 10) #создание строки из 10 равномерно заполненных элементов в диапозоне от 23 до 28, включая
e = np.diag([3,7,-2,5]) #создание диагональной матрицы с значениями из переданного массива в параметре 

print(a,b,c,d,e, sep='\n\n')
```

3. Демонстрация атрибутов:

```python
c.shape, c.dtype #массив с - одномерная массив (строка) из 12 элементов, тип данных масива c - float64
```

4. Примеры индексирования и срезов:
   - получение отдельного элемента;
   - извлечение подмассива с помощью среза.

```python
a[1,1] #получение отдельного элемента из двумерного массива из 2 строки и 2 столбца
```

```python
e[[2], 1:4] #получение подмассива 3 строки матрицы e длиною 3 с помощью среза, состоящего из элементов с индексами 1, 2 и 3
```

5. Примеры векторных операций и базовых статистик:
   - поэлементное сложение или умножение массивов;
   - вычисление средней величины (`mean`) или другой простой статистики.

```python
example1 = np.eye(2,3) #создание единичной матрицы 2 на 3
a * example1 #умножение единичной матрицы example1 на матрицу a
```

```python
example2 = np.vsplit(np.hsplit(e, [3])[0], 2)[0] #вертикально у горизонтпльное обрезание матрицы е до формы 2 на 3
a + example2 #сложение матрицы а и е
```

```python
#массив cars содержит данные о скорости(км/ч), расходе топлива(л/100км) и времени эксплуатации(лет), распределенные по столбцам соответсвтенно 
cars = np.array([
    [120, 8.5, 3],
    [150, 9.2, 5],
    [100, 7.8, 2],
    [130, 8.9, 4],
    [110, 8.0, 3]
])

average_features = cars.mean(axis=0)
print("Среднее по показателям (скорость, расход, возраст):", average_features)

feature_min = cars.min(axis=0)
print("Минимум по показателям:", feature_min)

feature_max = cars.max(axis=0)
print("Максимум по показателям:", feature_max)

feature_std = cars.std(axis=0)
print("Стандартное отклонение по показателям:", feature_std)
```
